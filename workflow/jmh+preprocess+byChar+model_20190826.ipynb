````python
act_game_money = train_activity.groupby(['acc_id', 'char_id', 'day'])['game_money_change'].sum()
act_game_money = act_game_money.unstack().fillna(0)

act_gm_change3 = act_game_money.T.rolling(window=3).sum().dropna()[::3].T
act_gm_change7 = act_game_money.T.rolling(window=7).sum().dropna()[::7].T
act_gm_change14 = act_game_money.T.rolling(window=14).sum().dropna()[::14].T

act_game_money.columns = ['game_money_change' + str(col) for col in act_game_money.columns]
act_gm_change3.columns = ['sum3_game_money_change' + str(col) for col in act_gm_change3.columns]
act_gm_change7.columns = ['sum7_game_money_change' + str(col) for col in act_gm_change7.columns]
act_gm_change14.columns = ['sum14_game_money_change' + str(col) for col in act_gm_change14.columns]
````

````python
act_game_money_ = pd.concat((act_game_money, act_gm_change3, act_gm_change7, act_gm_change14), axis=1)
````

````python
acc_char_info = acc_char_info.set_index(['acc_id', 'char_id'])
````

````python
act_game_money_ = act_game_money_.loc[acc_char_info.index]
````

````python
acc_char_info = pd.concat((acc_char_info, act_game_money_), axis=1)
````

# 차원 축소

- Charcter
    1. 모든 캐릭터를 고려, 통계값 추출
    2. 대표 캐릭터를 선정, 그 값을 대표값으로 사용
- 기타
    3. 기존에 있던 변수 추가

````python
# 1. 모든 캐릭터를 고려, 통계값을 추출

# 유저별 일일당 접속 캐릭터 수
acc_char_info[['day' + str(i) for i in range(1, 29)]].groupby(level=0).sum()
# 유저별 출석일수
acc_char_info[['day' + str(i) for i in range(1, 29)]].groupby(level=0).max().sum(axis=1)
# 유저별 일일당 플레이시간 (캐릭터 고려)
acc_char_info['daily_playtime'].groupby(level=0).sum()
# 유저별 마지막날 - 첫날 레벨의 다양한 통계량
acc_char_info['level_variation'].reset_index().groupby('acc_id').agg(
    {'level_variation' : ['mean', 'median', 'first', 'last', 'min', 'max']})
# 유저별 레벨 상승 / 하락 / 변동 존재 / 무변동 여부
t = train_combat.groupby(['acc_id', 'char_id', 'day', 'server']).agg({'level':'first'}).groupby(level=[0,1,2]).head(1)
level_nunique = t.reset_index().groupby(['acc_id', 'char_id']).agg({'level' : 'nunique'})
lev_variation = np.logical_and(
    (acc_char_info['level_variation'] == 0).values,
    (level_nunique > 1).values.ravel())
lev_static = np.logical_and(
    (acc_char_info['level_variation'] == 0).values,
    (level_nunique <= 1).values.ravel())
p = pd.concat((acc_char_info['level_variation'] > 0, acc_char_info['level_variation'] < 0), axis=1)
p['a'] = lev_static
p['b'] = lev_variation
# 유저별 일별 부활-사망 데이터의 평균, 합, 최대값 통계량
acc_char_info[['is_revive' + str(i) for i in range(1, 29)]].groupby(level=0).agg(['mean', 'sum', 'max'])
# 유저별 ewm(, 5) 처리를 한 일별 부활-사망 데이터의 평균, 합, 최대값 통계량
acc_char_info[['is_revive' + str(i) + '_ewm5' for i in range(4, 29, 3)]].groupby(level=0).agg(['mean', 'sum', 'max'])
# 유저별 시간 데이터 종합 통계량
acc_char_info[['playtime' + str(i) for i in range(1, 29)]].groupby(level=0).agg(['mean', 'sum', 'max'])
acc_char_info[['playtime_' + str(i) + '_ewm5' for i in range(4, 29, 3)]].groupby(level=0).agg(['mean', 'sum', 'max'])
acc_char_info[['fishing' + str(i) for i in range(1, 29)]].groupby(level=0).agg(['mean', 'sum', 'max'])
acc_char_info[['fishing_' + str(i) + '_ewm5' for i in range(4, 29, 3)]].groupby(level=0).agg(['mean', 'sum', 'max'])
acc_char_info[['private_shop' + str(i) for i in range(1, 29)]].groupby(level=0).agg(['mean', 'sum', 'max'])
acc_char_info[['private_shop_' + str(i) + '_ewm5' for i in range(4, 29, 3)]].groupby(level=0).agg(['mean', 'sum', 'max'])
# 유저별 해당일 보스 타격 빈도수
acc_char_info['touch_boss'].groupby(level=0).sum()
# 유저별 해당일 ewm(, 5)처리를 한 경험치 데이터의 평균, 합, 최대값, 최소값 통계량
acc_char_info[['exp_' + str(i) + '_ewm5' for i in range(4, 29, 3)]].groupby(level=0).agg(['mean', 'sum', 'max', 'min'])
# 유저별 combat data 종합 통계량
acc_char_info[['random_attacker_cnt{}ewm7'.format(i) for i in [7, 14, 21, 28]]].groupby(level=0).agg(['mean', 'sum', 'max', 'min'])
acc_char_info[['random_defender_cnt{}ewm7'.format(i) for i in [7, 14, 21, 28]]].groupby(level=0).agg(['mean', 'sum', 'max', 'min'])
acc_char_info[['temp_cnt{}ewm7'.format(i) for i in [7, 14, 21, 28]]].groupby(level=0).agg(['mean', 'sum', 'max', 'min'])
acc_char_info[['etc_cnt{}ewm7'.format(i) for i in [7, 14, 21, 28]]].groupby(level=0).agg(['mean', 'sum', 'max', 'min'])
acc_char_info[['num_opponent{}ewm7'.format(i) for i in [7, 14, 21, 28]]].groupby(level=0).agg(['mean', 'sum', 'max', 'min'])

````
