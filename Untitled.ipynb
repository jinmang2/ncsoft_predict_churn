```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(facecolor='w')
s = np.sin(2 * np.pi * .125 * np.arange(20))
plt.plot(s, 'ro-')
plt.xlim(-.5, 20.5)
plt.ylim(-1.1, 1.1)
plt.grid(True); plt.show()
```

```python
from scipy.linalg import toeplitz

S = np.fliplr(toeplitz(np.r_[s[-1], np.zeros(s.shape[0] - 2)], s[::-1]))
S[:5, :3]
```

```python
S.shape
```

# Many to One
- `Keras`에서 RNN을 사용하려면 입력 데이터는 (`nb_samples`, `timesteps`, `input_dim`) 크기를 가지는 ndim=3인 3차원 Tensor 형태여야 한다.
    - `nb_samples` : 자료의 수
    - `timesteps` : 순서열의 길이
    - `input_dim` : X 벡터으 ㅣ크기
- 위의 예제에서는 단일 시계열이므로 `input_dim`=1이고 3 스탭 크기의 순서열을 사용하므로 `timesteps`=3이며 자료의 수는 18개 이다.

```python
plt.figure(facecolor='w')
# s = np.sin(2 * np.pi * .125 * np.arange(20))
plt.plot(S[12], 'ro-')
plt.xlim(-.5, 20.5)
plt.ylim(-1.1, 1.1)
plt.grid(True); plt.show()
```

```python
plt.figure(facecolor='w')
# s = np.sin(2 * np.pi * .125 * np.arange(20))
plt.plot(S[:-1, :3].flatten(), 'ro-')
# plt.xlim(-.5, 20.5)
plt.ylim(-1.1, 1.1)
plt.grid(True); plt.show()
```

```python
X_train = S[:-1, :3][: ,: , np.newaxis]
y_train = S[:-1, 3]

X_train.shape, y_train.shape
```

```python
X_train[0], y_train[0]
```

```python
fig = plt.figure(figsize=(10, 6), facecolor='w')

ax1 = fig.add_subplot(2, 1, 1)
ax1.plot([0, 1, 2], X_train[0].flatten(), 'bo-', label='input sequence')
ax1.plot([3], y_train[0], 'ro', label='target')
ax1.set_xlim(-.5, 4.5)
ax1.set_ylim(-1.1, 1.1)
ax1.legend(); ax1.grid(True)
ax1.set_yticks([-1, -.5, 0, .5, 1])
ax1.set_title('Fist sample sequence')

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot([1, 2, 3], X_train[1].flatten(), 'bo-', label='input sequence')
ax2.plot([4], y_train[1], 'ro', label='target')
ax2.set_xlim(-.5, 4.5)
ax2.set_ylim(-1.1, 1.1)
ax2.legend(); ax2.grid(True)
ax2.set_yticks([-1, -.5, 0, .5, 1])
ax2.set_title('Second sample sequence')

plt.tight_layout()
plt.show()
```

### Keras의 SimpleRNN 클래스
- Keras에서 신경망 모형은 다음과 같은 순서로 만든다.
    1. `Sequential` 클래스 객체인 모형을 생성한다.
    2. `add` 메서드로 다양한 레이어를 추가한다.
    3. `compile` 메서드로 목적함수 및 최적화 방법을 지정한다.
    4. `fit` 메서드로 가중치를 계산한다.

```python
from keras import backend as K
K.clear_session()

from keras.models import Sequential
from keras.layers import SimpleRNN, Dense

np.random.seed(42)
model = Sequential()
model.add(SimpleRNN(10, input_shape=(3, 1)))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='sgd')
model.summary()
```

```python
plt.figure(figsize=(10, 6), facecolor='w')
plt.plot(y_train, 'ro-', label='target')
plt.plot(model.predict(X_train[:, :, :]), 'bs-', label='output')
plt.xlim([-.5, 20])
plt.ylim([-1.1, 1.1])
plt.legend(); plt.grid(True)
plt.title('After training')
plt.show()
```

```python
history = model.fit(X_train, y_train, epochs=100)
```

```python
plt.figure(figsize=(10, 6), facecolor='w')
plt.plot(y_train, 'ro-', label='target')
plt.plot(model.predict(X_train[:, :, :]), 'bs-', label='output')
plt.xlim([-.5, 20])
plt.ylim([-1.1, 1.1])
plt.legend(); plt.grid(True)
plt.title('After training')
plt.show()
```

# Many to Many


```python
from keras import backend as K
K.clear_session()

from keras.models import Sequential
from keras.layers import SimpleRNN, Dense, TimeDistributed

np.random.seed(42)
model2 = Sequential()
model2.add(SimpleRNN(10, input_shape=(3, 1), return_sequences=True))
model2.add(TimeDistributed(Dense(1, activation='linear')))
model2.compile(loss='mse', optimizer='sgd')
model2.summary()
```

```python
X_train = S[:-3, 0:3][:, :, np.newaxis]
y_train = S[:-3, 3:6][:, :, np.newaxis]
X_train.shape, y_train.shape
```

```python
fig = plt.figure(figsize=(10, 6), facecolor='w')

ax1 = fig.add_subplot(2, 1, 1)
ax1.plot([0, 1, 2], X_train[0].flatten(), 'bo-', label='input sequence')
ax1.plot([3, 4, 5], y_train[0], 'ro-', label='target sequence')
ax1.set_xlim(-.5, 6.5)
ax1.set_ylim(-1.1, 1.1)
ax1.legend(); ax1.grid(True)
ax1.set_yticks([-1, -.5, 0, .5, 1])
ax1.set_title('Fist sample sequence')

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot([1, 2, 3], X_train[1].flatten(), 'bo-', label='input sequence')
ax2.plot([4, 5, 6], y_train[1], 'ro-', label='target sequence')
ax2.set_xlim(-.5, 6.5)
ax2.set_ylim(-1.1, 1.1)
ax2.legend(); ax2.grid(True)
ax2.set_yticks([-1, -.5, 0, .5, 1])
ax2.set_title('Second sample sequence')

plt.tight_layout()
plt.show()
```

```python
history = model2.fit(X_train, y_train, epochs=100)
```

```python
fig = plt.figure(figsize=(10, 6), facecolor='w')

ax1 = fig.add_subplot(2, 1, 1)
ax1.plot([0, 1, 2], X_train[0].flatten(), 'bo-', label='input sequence')
ax1.plot([3, 4, 5], y_train[0], 'ro-', label='target sequence')
ax1.plot([3, 4, 5], model2.predict(X_train[0:1]).flatten(), 'gs-', label='target sequence')
ax1.set_xlim(-.5, 6.5)
ax1.set_ylim(-1.1, 1.1)
ax1.legend(); ax1.grid(True)
ax1.set_yticks([-1, -.5, 0, .5, 1])
ax1.set_title('Fist sample sequence')

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot([1, 2, 3], X_train[1].flatten(), 'bo-', label='input sequence')
ax2.plot([4, 5, 6], y_train[1], 'ro-', label='target sequence')
ax2.plot([4, 5, 6], model2.predict(X_train[1:2]).flatten(), 'gs-', label='target sequence')
ax2.set_xlim(-.5, 6.5)
ax2.set_ylim(-1.1, 1.1)
ax2.legend(); ax2.grid(True)
ax2.set_yticks([-1, -.5, 0, .5, 1])
ax2.set_title('Second sample sequence')

plt.tight_layout()
plt.show()
```
